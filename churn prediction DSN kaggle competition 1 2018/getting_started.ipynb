```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import sklearn as sk
import mglearn as mg
import warnings
warnings.filterwarnings("ignore")
```

```python
# Applying PCA to the cancer dataset
from sklearn.datasets import load_breast_cancer
load_cancer = load_breast_cancer()
```

```python
x, y = load_cancer.data, load_cancer.target
```

```python
fig, axes = plt.subplots(15,2, figsize = (10,20))
malignant = load_cancer.data[load_cancer.target == 0]
benign = load_cancer.data[load_cancer.target == 1]
ax = axes.ravel()



for i in range(30):
    _ , bins = np.histogram(load_cancer.data[:,1], bins = 50)
    ax[i].hist(malignant[:,i], bins = bins, color = 'b', alpha = .5)
    ax[i].hist(benign[:, i], bins = bins, color = 'r', alpha = .5)
    ax[i].set_title(load_cancer.feature_names[i])
    ax[i].set_yticks(())

    
fig.tight_layout()
plt.suptitle("Cancer_histogram")
plt.show()
```
